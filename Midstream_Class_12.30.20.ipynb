```python

from CorporateCredit import CorporateCredit, DB

# from Database import PySQLite
# import pandas as pd
# import sqlite3

#To-dos



    
table='test_table1'    



class MidstreamCredit(CorporateCredit):
    def __init__(self, name):
        super().__init__(name)
        
    

    headings = ["Name","Ticker","Revenue","VCost","FCost","EBITDA","Capex","CommonDistro","PfdDistro","AvgIntCost",
                 "TotalDebt","Cash","CrudePXSensitivity","NGPXSensitivity",
                "CrudeVolSensitivity","NGVolSensitivity"] #all quarterly
    
    
    def setCurrentFCF(self):
        #DB = DB object from PySQLite class
        Revenue=DB.GetRecord(table,'Name','Revenue',self.name)
        Rev_g=DB.GetRecord(table,'Name','RevG',self.name)
        VCost=DB.GetRecord(table,'Name','VCost',self.name)
        FCost=DB.GetRecord(table,'Name','FCost',self.name)
        Cash=DB.GetRecord(table,'Name','Cash',self.name)
        Capex=DB.GetRecord(table,'Name','Capex',self.name)
        Int_cost=DB.GetRecord(table,'Name','AvgIntCost',self.name)
        TotalDebt=DB.GetRecord(table,'Name','TotalDebt',self.name)
        Pfd=DB.GetRecord(table,'Name','PfdDistro',self.name)
        Common=DB.GetRecord(table,'Name','CommonDistro',self.name)
        
        
        Capex_rev = Capex/Revenue
        Interest = Int_cost*TotalDebt/4 #annualized rate
        

        
        
        self.CurrentFCF = {'rev': Revenue, 'rev_g': Rev_g, 'var_costs': VCost, 'f_costs': FCost, 'cash': Cash, 'capex/rev': Capex_rev,'interest':Interest,'pfd_distro':Pfd,'common_distro':Common,'WC/rev': 0, 'tax':0}
        return(self.CurrentFCF)
    
    
def panda_style(val):
    color = 'red' if val > 500 else 'black'
    return 'color: %s' % color

ETP=MidstreamCredit("ETP")
WES=MidstreamCredit("WES")

ETP.setCurrentFCF()
WES.setCurrentFCF()
ETP_Proj=ETP.FCF_Proj(10)
WES_Proj=WES.FCF_Proj(10)
# Proj=Proj.style.applymap(panda_style)
ETP_Proj = ETP_Proj.style.format("{:,.0f}")
WES_Proj = WES_Proj.style.format("{:,.0f}")


ETP_Proj
WES_Proj



   
    









```
